```python
# 소프트웨어전공 202284011 김연재

import gzip    # gz 압축파일을 읽기위해 사용
import numpy as np
import matplotlib.pyplot as plt
from collections import OrderedDict
from common.layers import *
from common.optimizer import Adam
from simple_convnet import SimpleConvNet    # CNN 소스코드 -  SimpleConvNet을 이용

# fashion mnist를 다운 받아서 사용
# 다운 받아둔 파일을 불러와 신경망 학습에 사용할 수 있도록 처리 (dataset.mnist를 참고함)
def load_local_fashion_mnist(normalize=True, flatten=True, one_hot_label=False):

    # 다운 받은 파일을 로컬에서 불러옴
    key_file = {
        'train_img': 'train-images-idx3-ubyte.gz',
        'train_label': 'train-labels-idx1-ubyte.gz',
        'test_img': 't10k-images-idx3-ubyte.gz',
        'test_label': 't10k-labels-idx1-ubyte.gz'
    }

    dataset = {}    # 데이터셋 저장을 위한 딕셔너리 초기화

    # 라벨 데이터 로드 함수 정의
    def _load_label(file_name):
        with gzip.open(file_name, 'rb') as f:    # gz 파일에서 라벨 데이터를 읽어오고
            labels = np.frombuffer(f.read(), np.uint8, offset=8)    # np.frombuffer를 사용해 바이트 데이터를 NumPy 배열로 변환한다.
        return labels

    # 이미지 데이터 로드 함수 정의
    def _load_img(file_name, num_images):
        with gzip.open(file_name, 'rb') as f:
            data = np.frombuffer(f.read(), np.uint8, offset=16)
        if flatten:    # flatten 플래그에 따라 이미지를 1차원(784) 또는 3차원(1, 28, 28) 배열로 변환한다.
            data = data.reshape(num_images, 784)
        else:
            data = data.reshape(num_images, 1, 28, 28)
        return data

    # 데이터 로드 및 전처리
    dataset['train_img'] = _load_img(key_file['train_img'], 60000)
    dataset['train_label'] = _load_label(key_file['train_label'])
    dataset['test_img'] = _load_img(key_file['test_img'], 10000)
    dataset['test_label'] = _load_label(key_file['test_label'])

    # 정규화
    if normalize:
        for key in ('train_img', 'test_img'):
            dataset[key] = dataset[key].astype(np.float32) / 255.0

    # 원-핫 인코딩
    if one_hot_label:
        dataset['train_label'] = _change_one_hot_label(dataset['train_label'])
        dataset['test_label'] = _change_one_hot_label(dataset['test_label'])

    # 데이터셋 반환
    return (dataset['train_img'], dataset['train_label']), (dataset['test_img'], dataset['test_label'])

# 원-핫 인코딩 함수 정의
def _change_one_hot_label(X):
    T = np.zeros((X.size, 10))
    for idx, row in enumerate(T):
        row[X[idx]] = 1
    return T
    
```

```python
# 이 코드는 Fashion MNIST 데이터셋을 로드하고, CNN을 초기화한 후, 미니배치 경사 하강법을 사용해
# Adam 옵티마이저로 여러 에폭 동안 신경망을 학습시킨다.

# Fashion MNIST 데이터 로드
(x_train, t_train), (x_test, t_test) = load_local_fashion_mnist(normalize=True, flatten=False, one_hot_label=True)

# 학습용 함수
def train_network(network, x_train, t_train, x_test, t_test, epochs=20, batch_size=100, optimizer='Adam'):
    train_loss_list = []
    train_acc_list = []
    test_acc_list = []

    iter_per_epoch = max(x_train.shape[0] / batch_size, 1)
    optimizer = Adam()

    # 학습 루프
    for epoch in range(epochs):    # 외부 루프는 에폭시를 실행
        for i in range(int(iter_per_epoch)):    # 내부 루프는 미니배치를 실행
            batch_mask = np.random.choice(x_train.shape[0], batch_size)
            x_batch = x_train[batch_mask]
            t_batch = t_train[batch_mask]

            grads = network.gradient(x_batch, t_batch)
            optimizer.update(network.params, grads)

            loss = network.loss(x_batch, t_batch)
            train_loss_list.append(loss)

        # 정확도 계산
        train_acc = network.accuracy(x_train, t_train)
        test_acc = network.accuracy(x_test, t_test)
        train_acc_list.append(train_acc)
        test_acc_list.append(test_acc)
        
        # 학습 진행 사항
        print(f"epoch {epoch + 1} / {epochs} | train acc: {train_acc:.4f} | test acc: {test_acc:.4f}")

    # 학습 손실, 학습 정확도, 테스트 정확도의 리스트를 반환
    return train_loss_list, train_acc_list, test_acc_list

# 네트워크 학습 및 평가
network = SimpleConvNet(input_dim=(1, 28, 28), conv_param={'filter_num': 30, 'filter_size': 5, 'pad': 0, 'stride': 1},
                        hidden_size=100, output_size=10, weight_init_std=0.01)
train_loss, train_acc, test_acc = train_network(network, x_train, t_train, x_test, t_test, epochs=20, batch_size=100)

# 그래프 그리기 (optimizer_compare_mnist를 참고함)
plt.figure(figsize=(12, 8))
plt.plot(train_acc, label='Training Accuracy')
plt.plot(test_acc, label='Test Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
